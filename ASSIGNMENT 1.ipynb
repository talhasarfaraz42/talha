#QUESTION 1

```python
text = input("Enter a string: ")
letters=0
digits=0
specials=0

for character in text:
    if character.isalpha():
        letters += 1
    elif character.isdigit():
           digits += 1
    else:
          specials += 1

print("Number of letters:",letters)
print("Number of digits:",digits)
print("Number of specials characters:" ,specials)
```

#QUESTION 3

```python
sentence = input("Enter a Sentence: ")

print("Length of sentence:", len(sentence))

uppercase_count= 0
lowercase_count= 0

for a in sentence:
    if a.isupper():
        uppercase_count += 1
    elif a.islower():
        lowercase_count += 1

print("Number of uppercase letters:", uppercase_count)
print("Number of lowercase letters:", lowercase_count)
```

#QUESTION 7

```python
start = int(input("Enter the start of the range: "))
end = int(input("Enter the end of the range: "))
print("Numbers with even sum of digits: ")

for a in range(start, end + 1):
    digit_sum = 0
    for b in str(a):
        digit_sum = int(b)

    if digit_sum % 2 == 0:
        print(f"{a} --> sum of digits: {digit_sum}")
```

#QUESTION 8

```python
text = input("Enter a string: ")
character = {}

for a in text:
    if a in character:
        character[a] += 1
    else:
        character[a] = 1
        
print("Character frequencies (sorted): ")
for b in sorted(character):
    print(f"{b} : {character[b]}")
```
